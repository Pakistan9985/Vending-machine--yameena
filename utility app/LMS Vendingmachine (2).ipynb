```python
items = {
    1: ("Chips", 2),
    2: ("Sandwich", 3),
    3: ("Cookie", 4),
    4: ("Water", 5),
    5: ("Fanta", 8),
    6: ("Ice Coffee", 7),
    7: ("Protien Bar", 4),
}

print("-- Welcome To The Vending Machine!--")
for i, item in items.items():
    print(f"{i}. {item[0]} - {item[1]} sar")

choice = int(input("Please Type the item number: "))
money = float(input("Please Insert Money: "))

name, price = items[choice]

if money >= price:
    print(f"Dispensing {name}...")
    print(f"Please Take The Item from the Slot, THANK YOU!")
    print(f"Change: {money - price:.2f} sar")
else:
    print("NOT ENOUGH CASH")
```
